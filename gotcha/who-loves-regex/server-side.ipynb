```kotlin
import java.util.concurrent.Executors
import kotlin.system.measureTimeMillis

// This combines multiple evil techniques
val regex = Regex("^([a-z]+)+([a-z]+)+@test\\.com$")
val exec = Executors.newVirtualThreadPerTaskExecutor()
while (true) {
    for (x in 100..140) {
        val input = "ab".repeat(x) + "d"
        exec.submit {
            val time = measureTimeMillis {
                regex.matches(input)
            }
            println("Length: ${input.length}, Time: ${time}ms")
        }
    }

    Thread.sleep(1000)

}
```
